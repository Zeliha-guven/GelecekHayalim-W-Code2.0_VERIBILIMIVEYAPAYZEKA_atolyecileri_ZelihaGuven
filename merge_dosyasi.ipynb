```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

### CSV OKUMA

```python
df = pd.read_csv("final_iklim_analizi.csv")
git = pd.read_csv("owid-energy-data.csv")
co = pd.read_csv("owid-co2-data.csv")
```

### DATALARIN ANALİZİ

```python
co["person_gdp"] = co["gdp"] / co["population"]
#Kişi başı gelir
```

```python
co.head()
```

```python
columns_to_keep = [
    'country',
    'year',
    "iso_code",
    'co2_per_capita',
    'cumulative_co2',
    'energy_per_capita',
    'share_global_co2',
    "person_gdp",
    "population"
]
# co veri setinden tutulacak kolonların listesi
```

```python
co = co[columns_to_keep]
```

```python
df.drop(["social", "water", "food"], inplace=True,axis=1)
# Sonucu etkilemeyecek olduğunu düşündüğüm kolonlar
```

```python
git = git[["iso_code","year","other_renewables_share_elec","country"]]
# Git data setinden alınacak kolonlar
```

```python
git = git[(git["year"]>=1995) & (git["year"]<2024)]
# Yılları eşitlemek için yaptığım filtreleme
```

```python
co = co[(co["year"]>=1995) & (co["year"]<2024)]
# Aynısını co veri seti için yaptım
```

```python
co.head()
```

```python
df["iso_code"] = df["country"]
# Primary key imiz iso_code oldugu için ad değişikliği yaptım
```

```python
df.drop("country",inplace=True,axis=1)
# burda da country değişkenini dropladım
```

```python
df.head()
```

```python
merged_df = df.merge(co, on=['iso_code', 'year'], how='left')
merged_df = merged_df.merge(git, on=['iso_code', 'year'], how='left')

# burda da 3 veri setini df'e göre birleştirdim iso_code ve year baz alarak
```

```python
merged_df.columns
```

```python
merged_df.drop("country_y",inplace=True,axis=1)
# Bu kolonu dropladım çünkü bu kolonla aynı başka bir kolon vardı
```

```python
merged_df.isnull().sum()
```

```python
merged_df.drop(["co2_per_capita_x","cumulative_co2_x"],inplace=True,axis=1)
# Bu kolonların bütün değerleri null geldiği için dropladım
```
